```python
#filling the missing values and drop column
data['Ratings'] = data['Ratings'].fillna(data['Ratings'].mean())
data['Reviews'] = data['Reviews'].fillna(data['Reviews'].mean())
data['DiscountOffer'] = data['Reviews'].fillna(data['Reviews'].mean())
data.drop('DiscountPrice (in Rs)', axis = 1 , inplace = True)
```

```python
#DIVIDE the data into input and output
#input(x)-represents URL  
#output(y)-represents product id
#var.iloc[row slicing,column slicing]
#In the column slicing place , if there is a ':', then array is 2 dimensional
x = data.iloc[:,0:1].values
x #.values converts my dataframe into an array
```
